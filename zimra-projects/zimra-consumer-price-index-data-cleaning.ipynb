```python
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
df = pd.read_excel("/Volumes/myfirst/default/zimra/CPI_USD_May- 24_email.xls",header=1)
df
```

change the column names 

```python
df.columns = ["Year","Month"] + df.iloc[0,2:].values.tolist()
df = df.drop(index=0)
df
```

drop rows with all null values and the weights row

```python
df = df.dropna(how="all").reset_index(drop=True)
df = df.drop(index=0)
df
```

fill null values in _year_  column by previous non null value

```python
df[["Year"]] = df[["Year"]].ffill()
df
```

```python
df["date"] = pd.to_datetime(df["Year"].astype(str)+"-"+df["Month"],format="%Y-%b") + MonthEnd(0)
df = df.drop(columns=["Year","Month"])
df
```

Replacing null values in the inflation rate with 0, missing because of lag not available

```python
df[df.columns[-3:-1].tolist()] = df[df.columns[-3:-1].tolist()].fillna(0)
df
```

Check if there are still null values

```python
df.isna().sum()
```
