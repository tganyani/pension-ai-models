```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np
```

```python
# %pip install tensorflow
```

```python
df = pd.read_csv("/Volumes/myfirst/default/zimra/zimra-cunsumer-price-index-2019-2024.csv")
df.head()
```

```python
scaler = MinMaxScaler()
df = df.set_index('date')
data = scaler.fit_transform(df[[df.columns[-2]]])
```

```python
dataset = keras.utils.timeseries_dataset_from_array(
    data=data[:-1],
    targets=data[1:, 0],  # predict inflation_rate
    sequence_length=12,
    batch_size=12
)
```

```python
model = Sequential([
    LSTM(128, return_sequences=True, activation="tanh" ,input_shape=(12, 1)),
    LSTM(64,return_sequences=True,activation="tanh" ),
    LSTM(32, return_sequences=True),
    LSTM(16,return_sequences=True),
    LSTM(8),
    Dense(1)
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
model.fit(dataset, epochs=100)
```

```python
lookback = 12
last_window = data[-lookback:]  
future_steps = 18
future_predictions = []
window = last_window.copy()
for _ in range(future_steps):
    pred = model.predict(window.reshape(1, lookback, 1), verbose=0)

    future_predictions.append(pred[0, 0])
    # update window (shift left, add new prediction)
    window = np.vstack([window[1:], pred])
future_predictions= np.array(future_predictions)
future_predictions = scaler.inverse_transform(future_predictions.reshape(-1, 1))
future_predictions.shape
```

```python
predictions = model.predict(dataset, verbose=0)
predictions = scaler.inverse_transform(predictions)
```

```python
future_dates = pd.date_range(
    start=df.index[-13],
    periods=18,
    freq='M'
)
future_dates
```

```python
import plotly.express as px
fig = px.line(
    df,
    x=df.index,
    y=df.columns[-2],
    title='Inflation rate',
)
fig.add_scatter(
    x=df.index[:48],
    y=predictions.flatten(),
    # mode='',
    name='predicted'
)
fig
fig.add_scatter(
    x=future_dates,
    y=future_predictions.flatten(),
    mode='lines',
    name='future predictions'
)
fig.show()
```
