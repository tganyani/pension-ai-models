Delete null date dates

saving the cleaned file as a csv

### Cleaning the data for 2025

```python
df = pd.read_excel("/Volumes/myfirst/default/zimra/2025 ANNUAL EXCHANGE AVERAGE RATE.xlsx")
df
```

```python

df = df.drop(index=list(range(3))+[7,11,15]+list(range(19,26))).reset_index(drop=True)
df.columns = ["month"]+list(df.iloc[0,1:].values.astype(int))
df = df.drop(index=0)
df["month"] =df["month"].str[:3]
df
```

```python
df_long = df.melt(id_vars="month",value_name="average exchange rate",var_name="day")
df_long["date"] = pd.to_datetime("2025-"+df_long["month"].astype(str)+"-"+df_long["day"].astype(str),format="%Y-%b-%d",errors="coerce")
df_long = df_long.dropna(subset="date")
df_long = df_long[["date","average exchange rate"]]
df_long = df_long.sort_values("date",ascending=True)
df_long.to_csv("/Volumes/myfirst/default/zimra/cleaned-2025-average-exchange-rate-for-income-tax-returns-purpose.csv",index=False)
```

```python
df1 = pd.read_csv("/Volumes/myfirst/default/zimra/cleaned-2024-average-exchange-rate-for-income-tax-returns-purpose.csv")
df2 = pd.read_csv("/Volumes/myfirst/default/zimra/cleaned-2025-average-exchange-rate-for-income-tax-returns-purpose.csv")
df = pd.concat([df1,df2],ignore_index=True)
df = df.rename(columns={"average exchange rate":"average-exchange-rate"})
df.to_csv("/Volumes/myfirst/default/zimra/cleaned-average-exchange-rate-for-income-tax-returns-purpose.csv",index=False)
df = pd.read_csv("/Volumes/myfirst/default/zimra/cleaned-average-exchange-rate-for-income-tax-returns-purpose.csv")
df
```
